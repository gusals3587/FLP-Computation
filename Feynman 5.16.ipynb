```python
"""
the problem
a certain charged particle moves in an electric and a magnetic field according to the equations
dv_x/dt = -2v_y
dv_y/dt = 1 + 2v_x
t = 0, x = 0, y = 0, v_x = 1.00, v_y = 0
"""
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
v = np.zeros((500, 2)) # first is x, second is y
p = np.zeros((500, 2)) # position
a = np.zeros((500, 2))
v[0][0] = 1
a[0][0] = -2*v[0][1] # a_x = -2*v_y
a[0][1] = 1 + 2*v[0][0] # a_y = 1 + 2*v_x
# the coordinate shape is (x, y)
epsilon = 0.1
```

```python
for i in range(1, 500):
    v[i][1] = (v[i - 1][1] + 1/2 * (a[i - 1][1] + 1 + 2*v[i - 1][0] + a[i - 1][0]*epsilon)*epsilon) / (1 + epsilon**2)
    # most of this is was made by a lot of algebra and substitution
    a[i][0] = -2*v[i][1]
    v[i][0] = v[i - 1][0] + 1/2 * (a[i - 1][0] + a[i][0]) * epsilon
    # just regular verlet, nothing fancy here
    a[i][1] = 1 + 2*v[i][0]
```

```python
for i in range(1, 500):
    p[i][0] = p[i - 1][0] + v[i - 1][0]*epsilon + 1/2*a[i - 1][0]*(epsilon**2)
    p[i][1] = p[i - 1][1] + v[i - 1][1]*epsilon + 1/2*a[i - 1][1]*(epsilon**2)
```

```python
plt.plot([x[0] for x in p], [y[1] for y in p])
plt.plot([x[0] for x in p[:10]], [y[1] for y in p[:10]])
```
